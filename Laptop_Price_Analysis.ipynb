```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('/content/laptop_prices (1).csv')

# Set conversion rate (update if needed)
conversion_rate = 90  # 1 Euro = 90 INR (example)

# Check column names
print(df.columns)

# Assuming 'Price' is the column with Euro prices (change if it's named differently)
df['Price_INR'] = df['Price_euros'] * conversion_rate

# Preview updated DataFrame
print(df[['Price_euros', 'Price_INR']].head())

# Save to a new CSV file
df.to_csv('/content/laptop_prices_in_inr.csv', index=False)
```
